```python
import pandas as pd
data = pd.read_csv('messy_IMDB_dataset.csv', encoding='latin-1', delimiter=';')
data = data.drop_duplicates()
data
```

```python
data.columns = data.columns.str.strip() #removes any leading or trailing spaces
```

```python
data = data.rename(columns={'IMBD title ID': 'Title ID', 'Original titlÊ': 'Title', 'Release year': 'Release date', 'Genrë¨': 'Genre'})
```

```python
data.head()
```

```python
print(data['Country'].unique())
```

```python
data['Country'] = data['Country'].str.replace('USA', 'US')
data['Country'] = data['Country'].str.replace('New Zesland', 'New Zealand')
data['Country'] = data['Country'].str.replace('New Zeland', 'New Zealand')
data['Country'] = data['Country'].str.replace('West ', '')
data['Country'] = data['Country'].str.strip('.1')
print(data['Country'].unique())
```

```python
data
```

```python
print(data['Content Rating'].unique())
```

```python
# Fill missing ratings
data['Content Rating'] = data['Content Rating'].fillna('Unrated')

# Standardize common variations
data['Content Rating'] = data['Content Rating'].replace({
    'Not Rated': 'Unrated',
    'NR': 'Unrated',      # Optional if present
    'Unrated': 'Unrated'
})
```

```python
print(data['Content Rating'].unique())
```

```python
data = data.drop(columns="Unnamed: 8")
data
```

```python
print(data['Income'].unique())
```

```python
data['Income'] = data['Income'].str.replace('$ ', '')
data['Income'] = data['Income'].str.lstrip()
data['Income'] = data['Income'].str.replace('o', '0')
data['Income'] = data['Income'].str.replace(',', '')
data
```

```python
print(data['Votes'].unique())
```

```python
data['Votes'] = data['Votes'].str.replace('.', '', regex=False).astype('Int64')
data
```

```python
print(data['Score'].unique())
```

```python
data['Score'] = data['Score'].str.strip('.')
data['Score'] = data['Score'].str.replace(r'[^\d.,]', '', regex=True)
data['Score'] = data['Score'].str.replace(',', '.')
data['Score'] = data['Score'].str.replace(r'\.{2,}', '.', regex=True)
data['Score'] = data['Score'].astype(float)
data
```

```python
print(data['Release date'].unique())
```

```python
data['date'] = data['Release date'] # copy all release dates to a new column named date
data['date'] = data['date'].str.replace(r'(\d{1,2})(st|nd|rd|th)', r'\1', regex=True) # remove ordinals
data['date'] = data['date'].str.replace(r'\b(The |of|on|, |year)\b', '', regex=True) #analyse and remove extra text that is not needed
data['date'] = data['date'].str.replace('marzo', 'March') # replace non-english with english words
data['date'] = data['date'].str.replace('/', '-') # replaces / with - to standardize the formats
print(data['date'].unique())
```

```python
data['date'] = data['date'].str.replace(r'^\d{1,2} \d{1,2} \d{4}$', lambda x: x.group().replace(' ', '-'), regex=True) # remove extra whitespaces
data['date'] = data['date'].str.replace(r'(?i)\b(January|Jan|February|Feb|March|Mar|April|Apr|May|June|Jun|July|Jul|August|Aug|September|Sep|October|Oct|November|Nov|December|Dec)\b', lambda x: f"{pd.to_datetime(x.group(0), format='%B' if len(x.group(0)) > 3 else '%b').month:02}", regex=True) # replace months in text to their corresponding month in numberical format
data['date'] = data['date'].str.replace('--', '-') # removes any extra hyphens that exist while replcing the whitespaces
data['date'] = data['date'].str.lstrip(' -') # removes any spaces or - at the start of dates
print(data['date'].unique())
```

```python
from dateutil import parser
def standardize_date(x):
    try:
        return parser.parse(x, dayfirst=True, yearfirst=False).strftime('%Y-%m-%d') # for standard format for the dates
    except:
        return x  # or return x if you want to keep the original value

data['date'] = data['date'].astype(str).apply(standardize_date)
data['date'] = pd.to_datetime(data['date'], errors='coerce')
data
```

```python
data['Title ID'] = data['Title ID'].astype(str)  # Assuming ID as string
data['Title'] = data['Title'].astype(str)
data['Genre'] = data['Genre'].astype(str)
data['Duration'] = pd.to_numeric(data['Duration'], errors='coerce')  # Converts to float, handles errors
data['Country'] = data['Country'].astype(str)
data['Content Rating'] = data['Content Rating'].astype(str)
data['Director'] = data['Director'].astype(str)
data['Income'] = pd.to_numeric(data['Income'], errors='coerce')  # In case it contains commas or strings
data['Votes'] = pd.to_numeric(data['Votes'], errors='coerce')
data['Score'] = pd.to_numeric(data['Score'], errors='coerce')
```

```python
genre_stats = (
    data.assign(Genre=data['Genre'].str.split(', '))
        .explode('Genre')
        .groupby('Genre')
        .agg({'Title': 'count', 'Score': 'mean', 'Income': 'mean'})
        .reset_index()
        .rename(columns={'Title': 'Number of Titles', 'Score': 'Average Score', 'Income': 'Average Income'})
)
genre_stats = genre_stats.sort_values(by='Average Score', ascending=False)
genre_stats = genre_stats.dropna()
genre_stats=genre_stats.reset_index(drop=True)
genre_stats['Average Score'] = genre_stats['Average Score'].round(2)
genre_stats
```

```python
rating_group = data.groupby('Content Rating').agg({
    'Title': 'count',
    'Score': 'mean',
    'Income': 'mean'
}).reset_index()
rating_group.columns = ['Content Rating', 'Number of Titles', 'Average Score', 'Average Income']
rating_group = rating_group.sort_values(by='Average Score', ascending=False)
rating_group['Average Score'] = rating_group['Average Score'].round(2)
rating_group['Average Income'] = rating_group['Average Income'].round(2)
rating_group.reset_index(drop=True, inplace=True)
rating_group
```
